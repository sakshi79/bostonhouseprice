```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
print(boston.DESCR)
```

```python
print(boston.data)
```

```python
print(boston.target)
```

```python
print(boston.feature_names)
```

```python
dataset = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price'] = boston.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.corr()
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
#Dropping certain features with larger correlation with other features or lesser with target
X.drop(['DIS', 'RAD', 'CHAS'], axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle
```

```python
pickle.dump(scaler, open('scaling.pkl', 'wb'))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.coef_)
```

```python
print(lm.intercept_)
```

```python
lm.get_params()
```

```python
pred = lm.predict(X_test)
```

```python
plt.scatter(y_test, pred)
```

```python
residuals = y_test - pred
```

```python
residuals
```

```python
sns.displot(residuals)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print("MSE: ", mean_squared_error(y_test, pred))
print("MAE: ", mean_absolute_error(y_test, pred))
print("R2 score: ", r2_score(y_test, pred))
```

```python
#Adjusted R2 score
1 - (1-r2_score(y_test, pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
pickle.dump(lm, open('house_model.pkl', 'wb'))
```
